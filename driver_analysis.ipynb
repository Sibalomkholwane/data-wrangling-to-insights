# Driver Data Analysis and Cleaning
This notebook performs data cleaning and exploratory data analysis (EDA) on the driver data set, focusing on distance traveled and speeding percentages.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Settings for plots
plt.style.use('ggplot')
%matplotlib inline
```

## 1. Loading and Initial Inspection
We first attempt to load the raw data and inspect its structure.

```python
# Load the raw data
df_raw = pd.read_csv('driver-data.csv')
print("Shape of raw data:", df_raw.shape)
print("\nFirst 5 rows of raw data:")
print(df_raw.head())
```

## 2. Data Cleaning
The dataset appears to be formatted as a single column containing comma-separated values. We need to split this into distinct columns and convert them to the correct data types.

```python
# Extract column names from the raw header
header = df_raw.columns[0].split(',')

# Split the values into separate columns
df = df_raw[df_raw.columns[0]].str.split(',', expand=True)
df.columns = header

# Convert columns to numerical types
df['id'] = pd.to_numeric(df['id'])
df['mean_dist_day'] = pd.to_numeric(df['mean_dist_day'])
df['mean_over_speed_perc'] = pd.to_numeric(df['mean_over_speed_perc'])

# Check for missing values and duplicates
print("Missing values:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())

print("\nCleaned Data Summary:")
print(df.info())
df.head()
```

## 3. Exploratory Data Analysis (EDA)
We analyze the distribution of driver metrics to find patterns.

```python
# Summary statistics
print(df.describe())

# Visualizing distributions
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['mean_dist_day'], kde=True, ax=ax[0])
ax[0].set_title('Distribution of Mean Distance per Day')

sns.histplot(df['mean_over_speed_perc'], kde=True, ax=ax[1])
ax[1].set_title('Distribution of Overspeeding Percentage')

plt.tight_layout()
plt.show()
```

### 4. Relationship Analysis
We look for a relationship between distance and speeding behavior.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mean_dist_day', y='mean_over_speed_perc', alpha=0.5)
plt.title('Mean Distance vs. Mean Overspeeding %')
plt.xlabel('Mean Distance Day')
plt.ylabel('Mean Over Speed %')
plt.show()
```

### 5. Driver Segmentation (Clustering)
Based on the scatter plot, we can see natural groupings. We'll use K-Means to segment the drivers.

```python
from sklearn.cluster import KMeans
# Using K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(df[['mean_dist_day', 'mean_over_speed_perc']])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mean_dist_day', y='mean_over_speed_perc', hue='cluster', palette='viridis', alpha=0.6)
plt.title('Driver Clusters')
plt.show()
```

## 6. Conclusions and Findings
1. The data was initially misformatted and required string splitting.
2. Most drivers cover shorter distances and have low speeding percentages.
3. There is a clear group of long-distance drivers who also tend to speed more frequently.
4. K-Means identified distinct groups that can be used for targeted driver safety programs.
